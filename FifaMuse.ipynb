```python
#First we need to install all necessary modules
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Then we need to download the csvs into the notebooks
fifa_path = '/content/drive/MyDrive/CIS4500-Final-Data/players_22.csv'
club_stats_path =  '/content/drive/MyDrive/CIS4500-Final-Data/club_league_data_combined.csv'
```

```python
#FIFA-DATA
fifa_col_names = [
  'sofifa_id', 
  'long_name',
  'player_positions',
  'value_eur',
  'overall',
  'wage_eur',
  'age',
  'dob',
  'height_cm',
  'weight_kg',
  'club_name',
  'league_name',
  'club_jersey_number',
  'nation_jersey_number',
  'nationality_name',
  'player_face_url',
  'club_logo_url',
  'club_flag_url',
  'nation_logo_url',
  'nation_flag_url'
]
fifa_players = pd.read_csv(
    fifa_path,
    index_col=False,
    usecols=fifa_col_names,
)
fifa_players.head()
#fifa_players.shape[0]
```

```python
#CLUB_STATS-DATA
club_stats_col_names = [
  'team_name', 
  'common_name',
  'country',
  'matches_played',
  'matches_played_home',
  'matches_played_away',
  'wins',
  'losses',
  'wins_home',
  'wins_away',
  'draws',
  'goals_scored',
  'goals_conceded',
  'goal_difference',
  'cards_total',
  'shots',
  'shots_on_target',
]
club_stats = pd.read_csv(
    club_stats_path,
    index_col=False,
    usecols=club_stats_col_names,
)
club_stats.rename(columns={'team_name': 'club_name'}, inplace=True)
club_stats.head()
#club_stats.shape[0]
```

```python
#Now we need to go through the data and remove players who have missing data
missing_fifa = fifa_players.isin([" ?"])
missing_fifa.head()
missing_fifa = missing_fifa.any(axis=1)
fifa_players = fifa_players.loc[(~missing_fifa).values, :]
fifa_players.isin([" ?"]).any().any()

#Remove all players who are not a part of a league
fifa_players = fifa_players[fifa_players['league_name'].notna()]
```

```python
# Handling Position
fifa_players['position'] = fifa_players['player_positions'].apply(lambda x: x.split(',')[0])
fifa_players.head(5)
```

```python
#Now we need to find the indexes for each table
#Let's check "sofifa_id" as our index
len(fifa_players["sofifa_id"].unique()) == len(fifa_players["sofifa_id"])

#It's true so we can use sofifa_id as our index
fifa_players.index.name = "sofifa_id"
```

```python
#Now we can finally convert the csv back as a cleaned csv
fifa_players.to_csv("/content/drive/MyDrive/CIS4500-Final-Data/fifa_players_cleaned.csv", index=True)

#Number of Rows
fifa_players.shape[0]
```

```python
#Club_Belongs_League
#Get a table with league, club_name, and league_country
leagues_and_clubs = fifa_players.loc[:, ["league_name", "club_name"]].drop_duplicates()
club_belongs_league_df = pd.merge(club_league_df, leagues_and_clubs, left_on='name', right_on='league_name')
club_belongs_league_df = club_belongs_league_df.drop('name', axis=1)
club_belongs_league_df.rename(columns={'country': 'league_country'}, inplace=True)
club_belongs_league_df.to_csv("/content/drive/MyDrive/CIS4500-Final-Data/club_belongs_league_cleaned.csv", index=False)
```
